3.Dari data0.csv, lakukan perhitungan setiap kolomnya dan simpan perhitungan setiap kolomnya di parameter dataset baru pada kolom yang sama:<br>
-Rata-rata<br>
-Standar deviasi<br>
-Skewness<br>
-Median<br>
-Energi<br>
-Min<br>
-Max

```python
parameters = pd.DataFrame(index=['Average', 'Standard deviation', 'Skewness', 'Median', 'Energy', 'Min', 'Max'])

parameter_dfs = []

for column in data.columns:
    avg = data[column].mean()
    std_dev = data[column].std()
    skewness = data[column].skew()
    median = data[column].median()
    energy = np.sum(data[column] ** 2) / len(data[column])
    min_val = data[column].min()
    max_val = data[column].max()

    parameter_df = pd.DataFrame({
        column: [avg, std_dev, skewness, median, energy, min_val, max_val]
    }, index=['Average', 'Standard deviation', 'Skewness', 'Median', 'Energy', 'Min', 'Max'])

   parameter_dfs.append(parameter_df)

parameters = pd.concat(parameter_dfs, axis=1)

parameters.head(10)
```

4.Plot sinyal diatas dari no. 2 dengan multiplot 2x3, sinyal yang diplot dapat dipilih dari 6 kolom secara bebas

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

axes = axes.flatten()

for i in range(len(axes)):
    random_column = np.random.choice(data.columns)

    axes[i].plot(data.index, data[random_column])
    axes[i].set_title(random_column)
    axes[i].set_xlabel('Index')
    axes[i].set_ylabel('Value')

plt.tight_layout()

plt.show()
```

5.Plot scatterplot dari hasil no. 3 untuk rata-rata, buat warna dan tanda yang berbeda untuk setiap 10 kolom, kolom 1-10 warna merah *, kolom 11-20 warna biru o, kolom 21-30 warna hijau v, dst (bebas). Dimana x adalah nilai rata-rata dan y adalah nilai standar deviasi.

```python
num_columns = parameters.shape[1]

averages = parameters.iloc[0]
std_devs = parameters.iloc[1]

num_groups = num_columns // 10

colors = plt.cm.viridis(np.linspace(0, 1, num_groups))
markers = ['o', 's', 'D', '^', 'v', 'p', '*'] * (num_groups // len(markers) + 1)

plt.figure(figsize=(10, 6))
for i in range(num_groups):
    start_col = i * 10
    end_col = (i + 1) * 10
    group_averages = averages[start_col:end_col]
    group_std_devs = std_devs[start_col:end_col]
    plt.scatter(group_averages, group_std_devs, c=[colors[i]], marker=markers[i], label=f'Columns {start_col}-{end_col}', alpha=0.5)

plt.xlabel('Average')
plt.ylabel('Standard deviation')
plt.title('Scatterplot of Average vs Standard deviation')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))

plt.grid(True)
plt.show()
```

6.Lakukan hal yang sama spt no. 5 untuk skewness dan median

```python
num_columns = parameters.shape[1]

_skewness = parameters.iloc[2]
_median = parameters.iloc[3]

num_groups = num_columns // 10

colors = plt.cm.viridis(np.linspace(0, 1, num_groups))
markers = ['o', 's', 'D', '^', 'v', 'p', '*'] * (num_groups // len(markers) + 1)

plt.figure(figsize=(10, 6))
for i in range(num_groups):
    start_col = i * 10
    end_col = (i + 1) * 10
    group_skewness = _skewness[start_col:end_col]
    group_median = _median[start_col:end_col]
    plt.scatter(group_skewness, group_median, c=[colors[i]], marker=markers[i], label=f'Columns {start_col}-{end_col}', alpha=0.5)

plt.xlabel('Skewness')
plt.ylabel('Median')
plt.title('Scatterplot of Skewness vs Median')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))

plt.grid(True)
plt.show()
```
